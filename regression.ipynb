```python
import numpy as np
import matplotlib.pyplot as plt
import random

def plot_prediction(x,y,w):  
    plt.plot(x[:,0],y, 'r') # training set - red
    plt.plot(x[:,0],x.dot(w), 'b') # w - blue
    plt.show()
    
    
def reduction_from_polynomial_to_linear(max_poly_deg,x,y):
    x = np.reshape(x,(1,x.shape[0])).T
    y = np.reshape(y,(1,y.shape[0])).T
    bias = np.ones((1, x.shape[0]))
    xs = bias.T
    for rank in range(max_poly_deg):
        xs = np.append(x**(rank+1),xs,axis=1)
    return xs,y
    

def normal_equation(max_poly_deg,x,y):
    x,y=reduction_from_polynomial_to_linear(max_poly_deg,x,y)
    A_inv = np.linalg.inv((x.T).dot(x))
    b = (x.T).dot(y)
    w = A_inv.dot(b)
    plot_prediction(x,y,w)
    return w


def gradient_descent(x, y,rate, precision, max_poly_deg):
    x,y = reduction_from_polynomial_to_linear(max_poly_deg,x,y)
    w = np.random.rand(max_poly_deg+1,1)
    m = len(y)
    A = (x.T).dot(x)
    b = (x.T).dot(y)
    gradient = (1/m)* (A.dot(w) - b) 
    while np.linalg.norm(gradient, 'fro') > precision:
        w = w -rate*gradient  
        gradient = (1/m)* (A.dot(w) - b) 
    plot_prediction(x,y,w)
    return w

rate = 0.001
precision = 0.001

# Data Set 1 
x=np.load('x_train_1.npy')
y=np.load('y_train_1.npy')
print("Data Set: 1")
print("Normal Equation - Polynomial degree: 1")
max_poly_deg =1
w=normal_equation(max_poly_deg,x,y)
print("Gradient Descent - Polynomial degree: 1")
w = gradient_descent(x,y,rate,precision,max_poly_deg)
print("Normal Equation - Polynomial degree: 2")
max_poly_deg =2
w=normal_equation(max_poly_deg,x,y)
print("Gradient Descent - Polynomial degree: 2")
w = gradient_descent(x,y,rate,precision,max_poly_deg)
print("Normal Equation - Polynomial degree: 3")
max_poly_deg =3
w=normal_equation(max_poly_deg,x,y)
print("Gradient Descent - Polynomial degree: 3")
w = gradient_descent(x,y,rate,precision,max_poly_deg)


# Data Set 2 

x=np.load('x_train_2.npy')
y=np.load('y_train_2.npy')
print("Data Set: 2")
print("Normal Equation - Polynomial degree: 1")
max_poly_deg =1
w=normal_equation(max_poly_deg,x,y)
print("Gradient Descent - Polynomial degree: 1")
w = gradient_descent(x,y,rate,precision,max_poly_deg)
print("Normal Equation - Polynomial degree: 2")
max_poly_deg =2
w=normal_equation(max_poly_deg,x,y)
print("Gradient Descent - Polynomial degree: 2")
w = gradient_descent(x,y,rate,precision,max_poly_deg)
print("Normal Equation - Polynomial degree: 3")
max_poly_deg =3
w=normal_equation(max_poly_deg,x,y)
print("Gradient Descent - Polynomial degree: 3")
w = gradient_descent(x,y,rate,precision,max_poly_deg)


# Data Set 3 
x=np.load('x_train_3.npy')
y=np.load('y_train_3.npy')
print("Data Set: 3")
print("Normal Equation - Polynomial degree: 1")
max_poly_deg =1
w=normal_equation(max_poly_deg,x,y)
print("Gradient Descent - Polynomial degree: 1")
w = gradient_descent(x,y,rate,precision,max_poly_deg)
print("Normal Equation - Polynomial degree: 2")
max_poly_deg =2
w=normal_equation(max_poly_deg,x,y)
print("Gradient Descent - Polynomial degree: 2")
w = gradient_descent(x,y,rate,precision,max_poly_deg)
print("Normal Equation - Polynomial degree: 3")
max_poly_deg =3
w=normal_equation(max_poly_deg,x,y)
print("Gradient Descent - Polynomial degree: 3")
w = gradient_descent(x,y,rate,precision,max_poly_deg)
```
